```python
# BALANCE ENQUIRY

def balance_enquiry():
    account_no = input("Enter account no: ").strip().upper()
    if not account_no:
        print("No account number entered.")
    else:
        account_data = account_info_data.get(account_no, {})
        if not account_data:
            print("No account found.")
        else:
            balance = account_data["balance"]
            print("Your balance amount : ", balance)

balance_enquiry()
```

```python
# DEPOSITE MONEY

def deposite_money():
    account_no = input("Enter account no: ").strip().upper()
    if not account_no:
        return print("No account number entered.")

    account_data = account_info_data.get(account_no, {})

    if not account_data:
        print(f"No account found for {account_no}")
    else:
        try:
            deposit_amount = float(input("Enter amount: ").strip())
        except ValueError:
            print("Invalid value!")
            return

        if deposit_amount <= 0:
            return print("Amount should be greater that 0.")

        account_data["balance"] += deposit_amount
        balance = account_data["balance"]
        print(
            f"Your new balance is: {balance}/-",
        )

deposite_money()
```

```python
# WITHDRAW MONEY

def withdraw_money():
    account_no = input("Enter account no: ").strip().upper()
    if not account_no:
        return print("No account number entered.")

    account_data = account_info_data.get(account_no, {})

    if not account_data:
        print(f"No account found for {account_no}")
    else:
        try:
            withdraw_amount = float(input("Enter amount: ").strip())
        except ValueError:
            print("Invalid value!")
            return

        if withdraw_amount <= 0:
            return print("Amount should be greater that 0.")

        balance = account_data["balance"]
        if balance < withdraw_amount:
            return print(
                f"You don't have enough balance! Available balance: {balance}/-"
            )

        account_data["balance"] -= withdraw_amount

        updated_balance = account_data["balance"]
        print(
            f"{withdraw_amount}/- withdrawn!\nYour current balance is: {updated_balance}/-",
        )

withdraw_money()
```

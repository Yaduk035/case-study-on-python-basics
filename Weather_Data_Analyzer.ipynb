```python
# Weather Data Analyzer

weather_data = [
    {"temperature": 30, "humidity": 65},
    {"temperature": 31, "humidity": 70},
    {"temperature": 29, "humidity": 80},
    {"temperature": 28, "humidity": 85},
    {"temperature": 32, "humidity": 60},
    {"temperature": 33, "humidity": 55},
    {"temperature": 30, "humidity": 75},
    {"temperature": 27, "humidity": 88},
    {"temperature": 26, "humidity": 90},
    {"temperature": 29, "humidity": 72},
    {"temperature": 31, "humidity": 68},
    {"temperature": 30, "humidity": 74},
    {"temperature": 28, "humidity": 82},
    {"temperature": 27, "humidity": 86},
    {"temperature": 32, "humidity": 60},
    {"temperature": 33, "humidity": 58},
    {"temperature": 31, "humidity": 66},
    {"temperature": 30, "humidity": 70},
    {"temperature": 29, "humidity": 78},
    {"temperature": 28, "humidity": 84},
    {"temperature": 27, "humidity": 89},
    {"temperature": 26, "humidity": 92},
    {"temperature": 30, "humidity": 73},
    {"temperature": 31, "humidity": 69},
    {"temperature": 32, "humidity": 64},
    {"temperature": 33, "humidity": 59},
    {"temperature": 29, "humidity": 77},
    {"temperature": 28, "humidity": 83},
    {"temperature": 27, "humidity": 87},
    {"temperature": 26, "humidity": 91}
]

def calculate_avg_temperature(data):
    temps = [day["temperature"] for day in data]
    return sum(temps) / len(temps)

def calculate_avg_humidity(data):
    hums = [day["humidity"] for day in data]
    return sum(hums) / len(hums)

def get_weather_condition(temp, humidity):
    if humidity > 85:
        return "Rainy"
    elif temp > 30 and humidity < 70:
        return "Sunny"
    elif temp < 27:
        return "Cloudy"
    else:
        return "Moderate"

day = 0
try:
  day = int(input("Enter a day number (1-30): "))
except ValueError:
  print('Invalid input!')

if day and 1 <= day <= 30:
    selected_day = weather_data[day - 1]
    temp = selected_day["temperature"]
    humidity = selected_day["humidity"]

    condition = get_weather_condition(temp, humidity)

    print("\nWeather Details")
    print("----------------")
    print(f"Day: {day}")
    print(f"Temperature: {temp}°C")
    print(f"Humidity: {humidity}%")
    print(f"Condition: {condition}")
else:
    print("Invalid day. Please enter a number between 1 and 30.")


avg_temp = calculate_avg_temperature(weather_data)
avg_humidity = calculate_avg_humidity(weather_data)

print("\nAverage Weather Summary")
print("-----------------------")
print(f"Average Temperature: {avg_temp:.2f}°C")
print(f"Average Humidity: {avg_humidity:.2f}%")
```
